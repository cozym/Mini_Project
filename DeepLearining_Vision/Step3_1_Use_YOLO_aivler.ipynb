#### 1) Label íŒŒì¼ì˜ í´ë˜ìŠ¤ ê°’ ë³€ê²½

```python
import glob
```

```python
others_path = "/content/datasets/others"

txt_files = glob.glob(f"{others_path}/**/*.txt", recursive=True)

txt_files[:5]
```

```python
# íŒŒì¼ ê²½ë¡œ ì„¤ì •
for file_path in txt_files:
  # íŒŒì¼ ì½ê¸° ë° ìˆ˜ì •
  with open(file_path, 'r') as file:
      lines = file.readlines()

  # ê° ì¤„ì„ ê³µë°± ê¸°ì¤€ìœ¼ë¡œ splití•˜ì—¬ ì²« ë²ˆì§¸ ì›ì†Œë¥¼ 0ìœ¼ë¡œ ìˆ˜ì •
  modified_lines = []
  for line in lines:
      words = line.split()
      if words:  # ì¤„ì— ì›ì†Œê°€ ìˆì„ ê²½ìš°
          words[0] = '1'  # ì²« ë²ˆì§¸ ì›ì†Œë¥¼ 1ìœ¼ë¡œ ìˆ˜ì •
      modified_lines.append(" ".join(words) + "\n")  # ë‹¤ì‹œ ê³µë°±ìœ¼ë¡œ í•©ì³ ì €ì¥

  # ìˆ˜ì •ëœ ë‚´ìš©ì„ ë‹¤ì‹œ íŒŒì¼ì— ì €ì¥
  with open(file_path, 'w') as file:
      file.writelines(modified_lines)
```

#### 2) ëª¨ë¸ì´ ìš”í•˜ëŠ” êµ¬ì¡°ì˜ í´ë” ìƒì„±

```python
import os
import shutil

# ì›ë³¸ í´ë” ê²½ë¡œ ì„¤ì •
src_folders = {
    "my_train_images": "datasets/my/train/images",
    "my_valid_images": "datasets/my/valid/images",
    "my_test_images": "datasets/my/test/images",
    "others_train_images": "datasets/others/train/images",
    "others_valid_images": "datasets/others/valid/images",
    "others_test_images": "datasets/others/test/images",

    "my_train_labels": "datasets/my/train/labels",
    "my_valid_labels": "datasets/my/valid/labels",
    "my_test_labels": "datasets/my/test/labels",
    "others_train_labels": "datasets/others/train/labels",
    "others_valid_labels": "datasets/others/valid/labels",
    "others_test_labels": "datasets/others/test/labels"
}

# ìƒˆ í´ë” ê²½ë¡œ ì„¤ì •
dest_base_images = "dataset/images"
dest_base_labels = "dataset/labels"
dest_paths = {
    "train_images": os.path.join(dest_base_images, "train"),
    "valid_images": os.path.join(dest_base_images, "valid"),
    "test_images": os.path.join(dest_base_images, "test"),
    "train_labels": os.path.join(dest_base_labels, "train"),
    "valid_labels": os.path.join(dest_base_labels, "valid"),
    "test_labels": os.path.join(dest_base_labels, "test")
}

# ìƒˆ í´ë” êµ¬ì¡° ìƒì„±
for path in dest_paths.values():
    os.makedirs(path, exist_ok=True)

# íŒŒì¼ ë³µì‚¬ í•¨ìˆ˜ ì •ì˜
def copy_files(src, dest):
    for file_name in os.listdir(src):
        src_file = os.path.join(src, file_name)
        dest_file = os.path.join(dest, file_name)
        shutil.copy2(src_file, dest_file)

# ì´ë¯¸ì§€ íŒŒì¼ ë³µì‚¬
copy_files(src_folders["my_train_images"], dest_paths["train_images"])
copy_files(src_folders["others_train_images"], dest_paths["train_images"])

copy_files(src_folders["my_valid_images"], dest_paths["valid_images"])
copy_files(src_folders["others_valid_images"], dest_paths["valid_images"])

copy_files(src_folders["my_test_images"], dest_paths["test_images"])
copy_files(src_folders["others_test_images"], dest_paths["test_images"])

# ë¼ë²¨ íŒŒì¼ ë³µì‚¬
copy_files(src_folders["my_train_labels"], dest_paths["train_labels"])
copy_files(src_folders["others_train_labels"], dest_paths["train_labels"])

copy_files(src_folders["my_valid_labels"], dest_paths["valid_labels"])
copy_files(src_folders["others_valid_labels"], dest_paths["valid_labels"])

copy_files(src_folders["my_test_labels"], dest_paths["test_labels"])
copy_files(src_folders["others_test_labels"], dest_paths["test_labels"])

print("ì´ë¯¸ì§€ì™€ ë¼ë²¨ íŒŒì¼ì´ dataset/images ë° dataset/labels í´ë” ì•„ë˜ë¡œ ì„±ê³µì ìœ¼ë¡œ ë³µì‚¬ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
