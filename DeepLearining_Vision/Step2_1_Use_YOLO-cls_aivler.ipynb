```python
import os, zipfile
```

```python
data_me = path + '/Datasets/Keras/images_myface2.zip'
data_others = path + '/Datasets/Keras/lfw-deepfunneled.zip'
```

```python
def extract_zip(zip_file, extract_folder):
  if not os.path.exists(extract_folder):
      os.makedirs(extract_folder)

  with zipfile.ZipFile(zip_file, 'r') as zip_ref:
      file_list = zip_ref.namelist()
      for f in file_list:
          if not f.endswith('/') and f.lower().endswith('.jpg'):
              file_name = os.path.basename(f)

              if not file_name.startswith('._'):
                  d_path = os.path.join(extract_folder, file_name)

                  with zip_ref.open(f, 'r') as in_file, open(d_path, 'wb') as out_file:
                      out_file.write(in_file.read())
```

### (2) 데이터셋 전처리

* **세부 요구사항**
    - 데이터셋을 전처리 합니다.
        - YOLO-cls 모델이 요구하는 폴더 구조를 만듭니다.
            1. Datasets라는 폴더를 생성합니다.
            2. Training set, Validation set, Test set(선택 사항) 각 데이터셋이 들어갈 폴더를 생성합니다.
            3. 각 데이터셋 폴더에 분류할 클래스의 이름을 가진 폴더를 생성합니다.
        - 폴더 구조에 맞게 데이터를 분배합니다.
    - 예시 코드에서 사용한 라이브러리
        - os, glob, random, shutil, numpy

#### 1) 모델이 요하는 구조의 폴더 생성

```python
def create_yolo_cls_folders(base_dir):
    # Datasets 폴더 생성
    os.makedirs(base_dir, exist_ok=True)

    # Training set, Validation set, Test set 폴더 생성
    for dataset_type in ['train', 'val', 'test']:
        dataset_path = os.path.join(base_dir, dataset_type)
        os.makedirs(dataset_path, exist_ok=True)

        # 클래스 이름을 가진 폴더 생성
        os.makedirs(os.path.join(dataset_path, 'me'), exist_ok=True)
        os.makedirs(os.path.join(dataset_path, 'others'), exist_ok=True)
```

```python
# 사용 예시
create_yolo_cls_folders('/content/Datasets')
```

#### 2) 각 폴더에 이미지 데이터 옮기기

```python
import glob
import random
import shutil
import numpy as np

def split_and_copy_images(me_folder, others_folder, base_dir, train_ratio=0.8, val_ratio=0.1):
    # 각 폴더에서 이미지 파일 리스트 가져오기
    me_images = glob.glob(os.path.join(me_folder, '*'))
    others_images = glob.glob(os.path.join(others_folder, '*'))

    # 파일 순서 섞기
    random.shuffle(me_images)
    random.shuffle(others_images)

    # 각 데이터셋에 할당할 개수 계산
    def get_split_indices(num_images):
        train_count = int(num_images * train_ratio)
        val_count = int(num_images * val_ratio)
        test_count = num_images - train_count - val_count
        return train_count, val_count, test_count

    # 이미지 분할 함수
    def split_images(images, class_name):
        train_count, val_count, test_count = get_split_indices(len(images))
        split_sets = {
            'train': images[:train_count],
            'val': images[train_count:train_count + val_count],
            'test': images[train_count + val_count:]
        }

        for set_name, imgs in split_sets.items():
            for img in imgs:
                dest_dir = os.path.join(base_dir, set_name, class_name)
                shutil.copy(img, dest_dir)

    # me와 others 이미지 분할 및 복사
    split_images(me_images, 'me')
    split_images(others_images, 'others')
```

```python
# 사용 예시
me_folder = '/content/myface'
others_folder = '/content/others'
base_dir = '/content/Datasets'

split_and_copy_images(me_folder, others_folder, base_dir)
```
